```python
import tensorflow as tf
tf.__version__
```

```python
import numpy as np
```

```python
x_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=np.float32)
y_data = np.array([[0],[0],[0],[1]], dtype=np.float32)
```

```python
print('x_data : \n', x_data)
print('x_data shape : \n', np.shape(x_data))
print('y_data : \n', y_data)
print('y_data shape : \n', np.shape(y_data))
```

```python
# Placeholder
X = tf.placeholder(tf.float32, shape=[None, 2])
Y = tf.placeholder(tf.float32, shape=[None, 1])
```

```python
# Weight, bias Variables
W = tf.Variable(tf.random_normal([2, 1], name='weight'))
b = tf.Variable(tf.random_normal([1], name='bias'))

# Hypothesis
hypothesis = tf.matmul(X, W) + b
```

```python
# Cost function: Mean Squared Error
cost = tf.reduce_mean(tf.square(y_data - hypothesis))
# Gradient Descent Method
train = tf.train.GradientDescentOptimizer(learning_rate=0.02).minimize(cost)
```

```python
predicted = tf.cast(hypothesis>0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32))
```

```python
with tf.Session() as sess:
  sess.run(tf.global_variables_initializer())

  for step in range(3001):
    cost_val, _ = sess.run([cost, train], feed_dict={X:x_data, Y:y_data})
    if step % 100 == 0:
      print('step : ', step, ' cost : ', cost_val)

  h, p, a = sess.run([hypothesis, predicted, accuracy], feed_dict={X:x_data, Y:y_data})
  print("\nHypothesis: ", h, "\nPredicted: ", p, "\nAccuracy: ", a)
```
